## Importing the necessary python libraries and the data set

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the dataset

```python
from google.colab import files
uploaded=files.upload()
```

```python
df=pd.read_csv("world_population.csv")
df.head()
```

## Sort the data

```python
top10 = df.sort_values(by='2015 Population', ascending=False).head(10)
```

## Visualizing data using a Bar graph

```python
plt.figure(figsize=(12,6))
sns.barplot(
x='Country/Territory',
y='2015 Population',
data=top10,
hue='Country/Territory',
palette='viridis',
dodge=False
)
plt.legend([],[], frameon=False)
plt.title('TOP 10 MOST POPULATED COUNTRIES (2015)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```

## Visualizing data using a Histogram

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Density (per km²)'], bins=20, kde=True)
plt.title('Distribution of Population Densities')
plt.xlabel('Population Density (per km²)')
plt.ylabel('Number of Countries')
plt.show()
```
